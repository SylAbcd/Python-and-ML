```python
def wstaw (znak, ktoryel, poz):
    lista=list(ktoryel)
    return "".join(lista[0:poz]+list(znak)+lista[poz:])

def dodajDoZbioru(zbior, znak, etap):
    zbiortmp=set()
    for i in range (etap): #w ilu miejscach umieszczamy
        for elem in zbior: # w ktorym elem zbioru
            newelem=wstaw (znak,elem,i)
            zbiortmp.add(newelem) # wstaw 'znak' do j-tego elementu zbioru na poz i
    return zbiortmp


#print(dodajDoZbioru(dodajDoZbioru({'1'},'2',2),'3',3))


#listaDoPermutacji=['1','2','3','4']
listaDoPermutacji=['a','b','c','d']


spermutowany={}
spermutowany=listaDoPermutacji[0]

listaDodawanych=listaDoPermutacji[1:]

wIleMiejsc=2

for i in listaDodawanych:
    zbiorPermTmp=dodajDoZbioru(spermutowany,i,wIleMiejsc)
    wIleMiejsc+=1
    spermutowany=zbiorPermTmp
    zbiorPermTmp={}
    
for i in enumerate(spermutowany):
    print(i[0]," ",i[1])
```

```python
def perm(k):
    if k==1:
        print (P)
    else:
        for i in range(k):
            P[k],P[i]= P[i],P[k]
            print("przed ",P, " i ",i," k ",k)
            perm(k-1)
            P[k],P[i]= P[i],P[k]
            print("po    ",P)
```

```python
def druk(tab, tab_size):
    for i in range (tab_size):
        print(tab[i])

def perm(tab, tab_size):
    if (tab_size == 1):
        druk(tab, 3)
        print("\n")
    else:
        for i in range(tab_size):
            tab[tab_size - 1],tab[i]=tab[i], tab[tab_size - 1]
            perm(tab, tab_size - 1)
            tab[i],tab[tab_size-1]=tab[tab_size-1],tab[i]

tab=[1,2,3]
perm(tab, 3)
```
